## 2. Homeostatic Model

### 2.1 Distance from the Wall

```python
def get_distance_wall(log_file):
    locations = pd.read_csv(log_file, sep=' ', header=None)
    locations.columns = ['posx', 'posy']

    # Transform Unity coordinates and rescale it.
    locations['xdist'] = locations['posx'] - 1000
    locations['ydist'] = locations['posy'] - 1000
    locations['ydist'] = locations['ydist'] / 7
    locations['xdist'] = locations['xdist'] / 7

    # Calculate the minimum distance to all 4 walls and save it
    mindist = np.zeros(locations.shape[0])
    for idx, row in locations.iterrows():
        a = abs(row['xdist'] - 1)
        b = abs(row['xdist'] + 1)
        c = abs(row['ydist'] - 1)
        d = abs(row['ydist'] + 1)
        mindist[idx] = min(a, b, c, d)
    return mindist
```

```python
# Load file with the avatar coordinates
log_file1 = os.path.join(base_path,'locations_%.1f_%.1f.log' %(alpha1, target1))
log_file2 = os.path.join(base_path,'locations_%.1f_%.1f.log' %(alpha2, target2))
log_file3 = os.path.join(base_path,'locations_%.1f_%.1f.log' %(alpha3, target3))
```

```python
mindist1 = get_distance_wall(log_file1)
mindist2 = get_distance_wall(log_file2)
mindist3 = get_distance_wall(log_file3)
```

```python
# Take the min value(= min distance)from any of the 4 walls
plt.plot(mindist1)
plt.plot(mindist2)
plt.plot(mindist3)
plt.xlabel('Time')
plt.ylabel('Min Distance')
plt.xlim([0, mindist1.shape[0]])
```

###  2.2 Mean activity (Synaptic Gating)

```python
# Get data from avatars
def get_simulated_data(pickle_path,):
    simulated_data = cPickle.load(open(pickle_path, 'rb'))
    
    # transform data into daframe
    data = pd.DataFrame(simulated_data)
    
    # get simulation length
    sim_len = data.shape[0]
    
    mean_activity = np.zeros(data.shape[0])
    std_activity = np.zeros(data.shape[0])
    mean_firing = np.zeros(data.shape[0])
    std_firing = np.zeros(data.shape[0])
    mean_w = np.zeros(data.shape[0])
    std_w = np.zeros(data.shape[0])
    
    for idx, row in data.iterrows():
        mean_activity[idx] = np.mean(row['Sx'][0, :, 0], axis=0)
        std_activity[idx] = np.std(row['Sx'][0, :, 0], axis=0)
        # mulitply by 1000 in order to have the firing rate in seconds and not ms
        mean_firing[idx] = np.mean(row['Hx'], axis=0) * 1000
        std_firing[idx] = np.std(row['Hx'], axis=0) * 1000
        mean_w[idx] = np.mean(row['w'], axis=0)
        std_w[idx] = np.std(row['w'], axis=0)
        
    return mean_firing, std_firing, mean_activity, std_activity, mean_w, std_w, sim_len
    
```

```python
# Load pickle file where firing rate and mean activity are stored
pickle_path1 = os.path.join(base_path, 'simulated_data_%.1f_%.1f.cpickle' %(alpha1, target1))
pickle_path2 = os.path.join(base_path, 'simulated_data_%.1f_%.1f.cpickle' %(alpha2, target2))
pickle_path3 = os.path.join(base_path, 'simulated_data_%.1f_%.1f.cpickle' %(alpha3, target3))
```

```python
mean_firing1, std_firing1, mean_activity1, std_activity1, mean_w1, std_w1, sim_len1 = get_simulated_data(pickle_path1)
mean_firing2, std_firing2, mean_activity2, std_activity2, mean_w2, std_w2, sim_len2 = get_simulated_data(pickle_path2)
mean_firing3, std_firing3, mean_activity3, std_activity3, mean_w3, std_w3, sim_len3 = get_simulated_data(pickle_path3)
```

Although Unity ran for 60s for all 3 models the number of updates is different. Get the minimal number of updates among all 3 models

```python
min_len = min(sim_len1, sim_len2, sim_len3)
```

```python
# plot Mean activity rate over time for the different avatars
plt.plot(mean_activity1)
plt.plot(mean_activity2)
plt.plot(mean_activity3)
plt.xlabel('Time')
plt.ylabel('Mean Activity')
plt.xlim([0, min_len])
```

```python
# plot Mean activity rate over time for the different avatars
plt.plot(std_activity1)
plt.plot(std_activity2)
plt.plot(std_activity3)
plt.xlabel('Time')
plt.ylabel('SD Activity')
plt.xlim([0, min_len])
```
