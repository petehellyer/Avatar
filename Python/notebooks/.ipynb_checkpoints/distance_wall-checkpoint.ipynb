# Distance from the Wall

```python
%matplotlib inline
import os
import seaborn as sb
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
# Load file with the avatar coordinates
base_path = os.getcwd()
log_file = os.path.join(base_path,'locations.log')
locations = pd.read_csv(log_file, sep=' ', header=None)
locations.columns = ['posx', 'posy']
```

```python
# Transform Unity coordinates and rescale it.
locations['xdist'] = locations['posx'] - 1000
locations['ydist'] = locations['posy'] - 1000
locations['ydist'] = locations['ydist'] / 7
locations['xdist'] = locations['xdist'] / 7

# Calculate the minimum distance to all 4 walls and save it
mindist = np.zeros(locations.shape[0])
for idx, row in locations.iterrows():
    a = abs(row['xdist'] - 1)
    b = abs(row['xdist'] + 1)
    c = abs(row['ydist'] - 1)
    d = abs(row['ydist'] + 1)
    mindist[idx] = min(a, b, c, d)
```

```python
# Take the min value(= min distance)from any of the 4 walls
sb.tsplot(mindist, time='time',
plt.xlabel='time'
plt.ylabel='Min Distance'
```
