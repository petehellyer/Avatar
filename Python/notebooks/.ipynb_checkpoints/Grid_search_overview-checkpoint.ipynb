# Hypothesis 1:

How do the dynamics of the system look like if it learns from the synaptic gating.
Here, Homeostatic plasticity is defined as:

$$ w_{tmp} = alpha * (-w * (SG - target))$$

$$ w = w_{tmp} + w$$

where *SG* represents the synaptic gating

Plot Synchrony and Metastability for different G in the following order:
    - G: .1 below critical point
    - G: .198 Critical point
